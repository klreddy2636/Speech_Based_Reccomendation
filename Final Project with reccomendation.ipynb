# Real time audio recording

```python
CHUNK = 1024  # Chunk size in samples
FORMAT = pyaudio.paInt16  # Sample format
CHANNELS = 1  # Number of channels (mono)
RATE = 44100  # Sampling rate (Hz)

p = pyaudio.PyAudio()emb

stream = p.open(format=FORMAT,
                channels=CHANNELS,
                rate=RATE,
                input=True,
                frames_per_buffer=CHUNK)

frames = []  # Store audio chunks

print("Start recording...")

try:
    while True:
        data = stream.read(CHUNK)
        frames.append(data)
    # Implement logic for stopping recording (e.g., user input)
except KeyboardInterrupt:
    pass

print("Recording stopped.")

stream.stop_stream()
stream.close()
p.terminate()
```

# Saving the Recorded audio locally

```python


audio_data = b''.join(frames)
audio_data = np.frombuffer(audio_data, dtype=np.int16).reshape(-1, 1)  # Reshape for mono
print("Format: ",type(audio_data))
print("Shape",audio_data.shape)
soundfile.write("output.wav", audio_data, RATE, format="wav")  # Replace format if needed (e.g., "flac", "pcm")
print("Audio stored to output.wav")
```
